```python
# Cell 2 — Read all 20 CSVs into a single Spark DataFrame

base_path = "/Volumes/workspace/default/hmda_raw"

df_raw = (
    spark.read
    .format("csv")
    .option("header", "true")        # first row is column names
    .option("inferSchema", "false")  # we supply our own schema
    .schema(hmda_schema)
    .option("nullValue", "NA")       # HMDA uses "NA" for missing values
    .option("nullValue", "Exempt")   # some fields use "Exempt" too
    .option("mode", "PERMISSIVE")    # don't crash on bad rows, just nullify them
    .load([
        f"{base_path}/2023",
        f"{base_path}/2024"
    ])
)

print(f"DataFrame loaded")
print(f"   Rows    : {df_raw.count():,}")
print(f"   Columns : {len(df_raw.columns)}")
```

```python
# Cell 3 — Sanity checks on the raw data

# 1. Check row counts per year
print("=== Row counts by year ===")
df_raw.groupBy("activity_year").count().orderBy("activity_year").show()

# 2. Check row counts per state
print("=== Row counts by state ===")
df_raw.groupBy("state_code").count().orderBy("count", ascending=False).show()

# 3. Peek at a few rows
print("=== Sample rows (key columns only) ===")
df_raw.select(
    "activity_year",
    "state_code",
    "derived_race",
    "derived_ethnicity",
    "action_taken",
    "loan_amount",
    "interest_rate",
    "rate_spread"
).show(5, truncate=False)

# 4. Print schema to confirm types
print("=== Schema (key columns) ===")
df_raw.select(
    "activity_year",
    "state_code", 
    "derived_race",
    "action_taken",
    "loan_amount",
    "interest_rate",
    "rate_spread"
).printSchema()
```

```python
# Cell 4 — Null analysis on our most important columns
# These are the fields that drive our 3 research questions

from pyspark.sql.functions import col, count, when, round as spark_round

key_cols = [
    "activity_year",
    "state_code",
    "derived_race",
    "derived_ethnicity",
    "action_taken",
    "loan_amount",
    "interest_rate",
    "rate_spread",
    "income",
    "debt_to_income_ratio",
    "loan_type",
    "lien_status",
    "lei"          # lender identifier
]

total = df_raw.count()

null_counts = df_raw.select([
    spark_round(
        count(when(col(c).isNull(), c)) / total * 100, 2
    ).alias(c)
    for c in key_cols
])

print("=== Null % for key columns ===")
null_counts.show(truncate=False)
```

```python
# Cell 5 — Save raw data as Delta table (checkpoint)
# This is our "raw" layer in the medallion architecture:
# Raw (Delta) → Cleaned (Delta) → Aggregated (Delta)

delta_path = "/Volumes/workspace/default/hmda_raw/delta/raw_hmda"

(
    df_raw.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(delta_path)
)

print(f"Raw Delta table saved to {delta_path}")

# Read it back to confirm
df_check = spark.read.format("delta").load(delta_path)
print(f"   Rows    : {df_check.count():,}")
print(f"   Columns : {len(df_check.columns)}")
```
